```python
import requests
from bs4 import BeautifulSoup
import json
import datetime
import re
import time
import pdfkit
```

```python
def grab_links(url):
    link_list = []
    
    source_code = requests.get(url)
    plain_text = source_code.text
    soup = BeautifulSoup(plain_text, 'lxml')
    
#     for link in soup.findAll('a', attrs={'href': re.compile("^http://")}):
#     print link.get('href')
    
    for link in soup.find_all('a', attrs={'href': re.compile('.html$')}):
        link_list.append(link.get('href'))
        
    return link_list
```

```python
top_level_links = grab_links('https://otexts.org/fpp2/')
top_level_links
```

```python
def forecasting_spider(max_pages):
    page = 1
    
    start_time = time.time()

    for link in top_level_links:
        url = 'https://otexts.org/fpp2/' + link
        
        pdfkit.from_url(url, 'out' + str(page) + '.pdf')
        
#         # Use requests to return the html, but formatting is lost
#         r = requests.get(url)
#         with open("output" + str(page) + ".html", "wb") as file:
#             file.write(r.content)
        
        
#         source_code = requests.get(url)
#         plain_text = source_code.text
#         soup = BeautifulSoup(plain_text, 'html.parser')

#         html = soup.prettify("utf-8")
#         with open("output" + str(page) + ".html", "wb") as file:
#             file.write(html)

        page += 1

        if page >= max_pages:
            break
        elif page % 10 == 0:
            print('Finished {} pages in {:.2f} seconds'.format(page, (time.time() - start_time) / 60))

        time.sleep(1)
   
    print("done!")
```

```python
forecasting_spider(3)
```
